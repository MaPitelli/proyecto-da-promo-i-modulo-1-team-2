```python
import random
import os

class Ahorcado:
    def __init__(self):

        self.palabras_ahorcado = [
            "python", "programacion", "desarrollo", "algoritmo", "computadora",
            "tecnologia", "internet", "software", "hardware", "inteligencia",
            "artificial", "robotica", "electronica", "ingenieria", "matematica",
            "estadistica", "física", "quimica", "biologia", "genetica",
            "medicina", "astronomia", "geografia", "historia", "filosofia",
            "literatura", "arte", "musica", "pintura", "escultura",
            "arquitectura", "dibujo", "fotografia", "cine", "teatro",
            "danza", "poesia", "novela", "cuento", "ensayo",
            "periodismo", "redaccion", "publicidad", "marketing", "comunicacion",
            "educacion", "pedagogia", "psicologia", "sociologia", "antropologia"
        ]

        self.dibujo_ahoracado = list('''   +---+
        |   |
            |
            |
            |
            |
        =========

        ''')

        self.vidas = 6
        self.letras_falladas = []

    def punto_partido (self):
        palabra = random.choice(self.palabras_ahorcado)
        self.secreto = list("_" * len(palabra))
        os.system("clear")
        print(f"Bienvenido al ahorcado!!!\nLa palabra oculta contiene {len(palabra)} letras {' '.join(self.secreto)}\n\n\n") 
        print("".join(self.dibujo_ahoracado)) 
    
    def jugar (self):
        while self.vidas > 0:        
            letra_escogida = input(f"Escoge una letra ")
            if letra_escogida in self.palabra:
                for indices, letra in enumerate(self.palabra):    
                    if letra == letra_escogida:
                        self.secreto[indices] = letra_escogida 
                    os.system("clear")                      
                    print(f"Felicidades tu letra está en la palabra oculta {' '.join(self.secreto)}") 
                    print("".join(self.dibujo_ahoracado))             
                if self.palabra == "".join(self.secreto):
                    print(f"FELICIDADES has acertado la palabra {''.join(self.secreto)}, te sobraron {self.vidas} vidas") 
                    break   

    def comprobar_vidas (self):     
        if self.letra_escogida not in self.palabra:
            vidas -= 1
            self.letras_falladas.append(self.letra_escogida)
            os.system("clear")
            print(f"Lo siento, la palabra oculta no contiene la letra {self.letra_escogida}, las letras falladas son {self.letras_falladas}. \nSigue intentándolo {' '.join(self.secreto)} ")
            if vidas == 5:        
                self.dibujo_ahoracado[21] = "O"
                print("".join(self.dibujo_ahoracado))         
            elif vidas == 4:
                self.dibujo_ahoracado[30] = "|"
                print("".join(self.dibujo_ahoracado))           
            elif vidas == 3:
                self.dibujo_ahoracado[29] = "/"
                print("".join(self.dibujo_ahoracado))
            elif vidas == 2:
                self.dibujo_ahoracado[31] = "\\"
                print("".join(self.dibujo_ahoracado))
            elif vidas == 1:
                self.dibujo_ahoracado[38] = "/"
                print("".join(self.dibujo_ahoracado))
            elif vidas == 0:  
                self.dibujo_ahoracado[40] = "\\"       
                print("".join(self.dibujo_ahoracado))  
        else:
            print(f"Lo siento. Te has ahorcado la palabra era {self.palabra}")
    
```
